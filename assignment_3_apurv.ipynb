1. Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2. Read The Dataset

```python
student = pd.read_csv('student_graphs.csv')
student
```

3. Scatter Plot

#### Purpose: Relationship between two numerical variables (Maths vs Science)

```python
plt.scatter(student['Maths'], student['Science'])
plt.xlabel('Maths Marks')
plt.ylabel('Science Marks')
plt.title('Scatter Plot: Maths vs Science')
plt.show()
```

4. Bar Plot

#### Purpose: Compare marks of students in Maths

```python
plt.figure(figsize=(15, 6))
plt.bar(student['Student'], student['Maths'])
plt.xlabel('Students')
plt.ylabel('Maths Marks')
plt.title('Bar Plot: Maths Marks of Students')
plt.show()
```

5. Box Plot

#### Purpose: Show distribution and outliers of marks

```python
plt.boxplot([student['Maths'], student['Science'], student['English']],
            tick_labels=['Maths', 'Science', 'English'])
plt.title('Box Plot of Subject Marks')
plt.ylabel('Marks')
plt.show()
```

6. Pie Chart

#### Purpose: Percentage contribution of total marks by subject (average)

```python
average_marks = student[['Maths', 'Science', 'English']].mean()

plt.pie(average_marks, labels=average_marks.index, autopct='%1.1f%%')
plt.title('Pie Chart: Average Marks Distribution')
plt.show()
```

7. Line Chart

#### Purpose: Trend of marks across students

```python
plt.figure(figsize=(15, 6))
plt.plot(student['Student'], student['Maths'], label='Maths')
plt.plot(student['Student'], student['Science'], label='Science')
plt.plot(student['Student'], student['English'], label='English')

plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Line Chart: Marks Trend')
plt.legend()
plt.show()
```

8. Histogram

#### Purpose: Marks Distribution

```python
plt.hist(student['Maths'], bins=10, color='skyblue', edgecolor='black')
plt.title('Maths Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
```

```python
plt.hist(student['English'], bins=10, color='red', edgecolor='black')
plt.title('English Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
```

```python
plt.hist(student['Science'], bins=10, color='yellow', edgecolor='black')
plt.title('Science Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
```

9. Correlation Matrix

```python
corr = student[['Maths', 'Science', 'English']].corr()
plt.matshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.show()
```

10. Linear Regression - Actual VS Predicted

```python
from sklearn.linear_model import LinearRegression
X, y = student[['Maths', 'Science']], student['English']
student['Pred'] = LinearRegression().fit(X, y).predict(X)
student.head(10).plot(x='Student', y=['English', 'Pred'], kind='bar')
plt.title('Actual vs Predicted Scores'); plt.ylabel('Score')
plt.show()
```
