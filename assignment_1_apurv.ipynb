### Implementation

1. Import Required Libraries

```python
import pandas as pd
import numpy as np
```

2. Create Dataset and DataFrame

```python
data = {
    "Student_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Apurv", "Viraj", "Sahil", "Nisha", "Vaishnavi", "Aditya"],
    "Age": [21, 22, 20, 23, 21, 22],
    "Gender": ["M", "M", "M", "F", "F", "M"],
    "Department": ["CSAI", "CSE", "CSE", "CSAI", "AIDS", "ECE"],
    "Marks": [85, 90, np.nan, 78, 88, 0],
    "Attendance": [90, 95, 85, 80, 92, 0],
    "Internship": ["Yes", "No", "Yes", "No", "Yes", "No"]
}

df = pd.DataFrame(data)
df
```

3. Finding Shape of DataFrame

```python
print(df.shape)
```

4. Finding Missing Values in Each Column

```python
print(df.isnull().sum())
```

5. Fetching Data Types of Each Column

```python
print(df.dtypes)
```

6. Counting Number of zero values

```python
print((df == 0).sum())
```

7. Selecting First 2 Rows from dataset

```python
print(df.iloc[:2])
```

8. Selecting Name & Marks Columns from dataset

```python
print(df[['Name', 'Marks']])
```

9. Sorting by Marks


```python
print(df.sort_values(by="Marks"))
```

10. Describe Attributes of Data

```python
print(df.describe(include='all'))
```

11. Rechecking Data Types

```python
print(df.info())
```

12. Counting Unique Values


```python
print(df.nunique())
```

13. Fetching Column Names

```python
print(df.columns)
```

14. Converting Datatypes

```python
df['Student_ID'] = df['Student_ID'].astype('int16')
df['Age'] = df['Age'].astype('int8')
df['Marks'] = df['Marks'].astype('float16')
print("Data Types After Converting:")
print(df.dtypes)
```
